```python
def my_range(start,end):
    current = start
    while current < end:
        yield current
        current += 1
        
```

```python
t = my_range(4,8)
```

```python
#define a Counter class as an example
class Counter:
    def __init__(self,start,end):
        self.num = start
        self.end = end

    def __iter__(self):
        return self


    def __next__(self):
        if self.num > self.end:
            raise StopIteration
        current = self.num
        self.num += 1
        return current
```

```python
c1 = Counter(2,5)
```

```python
for i in c1:
    print(i)
```

```python
#Print with iter()
c2 = Counter(5,10)
myiter = iter(c2)
try:
    while True:
        print(next(myiter))
except:
    print("end of counting")
```

```python
#Decorators
def decorator(func):
    def wrapper():
        print("Before calling the function")
        func()
        print("After calling the function")
    return wrapper
```

```python
@decorator
def greet():
    print("Hello, Virginia")
    
```

```python

def decorator(func):
    def wrapper(*args):
        print("before...")
        func(*args)
        print("after...")
    return wrapper
    

def cal(x,y):
    print(x+y)
```

```python
cal(2,5)
```

```python
@decorator
def cal(x,y):
    print(x+y)
cal(2,3)
```

```python
@decorator
def cal(x,y,z):
    print(x*y+z)
cal(2,3,4)
```

```python
def my_decorator(func):
    def wrapper(*args,**kwargs):
        print("before...")
        result=func(*args,**kwargs)
        print(f"after...{result}")
        
    return wrapper
```

```python
@my_decorator
def add(m,n):
    
    return m + n
   
add(2,5)
```

```python
#Method Decorators
```

```python
def method_dec(func):
    def wrapper(self,*args,**kwargs):
        print("Before method execution")
        res=func(self, *args, **kwargs)
        print("After mehod execution")
        return res
    return wrapper

class MyClass:
    @method_dec
    def say_hello(self):
        print("Hello")
        
    @method_dec
    def mul(self,a,b):
        print (a*b)
        
obj = MyClass()
obj.say_hello()
obj.mul(3,4)
```
